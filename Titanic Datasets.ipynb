```python
sns.heatmap(loan_datasets.isnull())
```

```python
loan_datasets.head()
```

```python
loan_datasets.drop(["Loan_ID","Loan_Amount_Term"], axis=1, inplace=True)
```

```python
loan_datasets.head()
```

```python
Gender= pd.get_dummies(loan_datasets["Gender"],drop_first=True)
Married= pd.get_dummies(loan_datasets["Married"],drop_first=True)
Education= pd.get_dummies(loan_datasets["Education"],drop_first=True)
Self_Employed= pd.get_dummies(loan_datasets["Self_Employed"],drop_first=True)
Property_Area= pd.get_dummies(loan_datasets["Property_Area"],drop_first=True)
Loan_Status = pd.get_dummies(loan_datasets["Loan_Status"],drop_first=True)
```

```python
Dependents = pd.get_dummies(loan_datasets["Dependents"],drop_first=True)
```

```python
loan_datasets = pd.concat([Gender,Married,Education,Self_Employed,Property_Area,Loan_Status],axis= 1)
```

```python
x =loan_datasets.drop(['Y'],axis=1)
y = loan_datasets['Y'] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=23)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logimodel = LogisticRegression(random_state=23)
```

```python
logimodel.fit(x_train,y_train)
```

```python
predictions = logimodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)*100
```
